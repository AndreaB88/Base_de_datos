```python
#Ejercicio 2: Realiza las tareas de preprocesado y adecuación del Dataset que disponemos en el repositorio de GitHub
#PRE-PROCESSING-FECHA con el objetivo de prepararlo y trabajarlo como dataframe para extraer información.

import pandas as pd
import matplotlib.pyplot as plt
import requests


url = "https://raw.githubusercontent.com/BIDDL3/PRE-PROCESSING-DATA/main/movies.dat"
response = requests.get(url)
data = response.text 

```

```python
lines = data.split('\n')

movie_names = []
movie_years = []
movie_genres = []

for line in lines:
    elements = line.split("::")
    
    if len(elements) == 3:
    
        movie_name = elements[1]
        movie_year = int (elements[1][-5:-1]) if elements[1][-5:-1].isdigit() else None
        movie_genre = elements[2]
    
        movie_names.append(movie_name)
        movie_years.append(movie_year)
        movie_genres.append(movie_genre)
    
df = pd.DataFrame({'Nombre de la Película':movie_names, 'Año': movie_years, 'Género' : movie_genres})

df.drop_duplicates(subset='Nombre de la Película', keep='first', inplace=True)


df['Género'] = df['Género'].apply(lambda x: x.split('|')[0] if '|' in x else x)

df.reset_index(drop=True, inplace=True)

print(df.head(10))

```

```python
import re

def limpiar_año(text):
    return re.sub(r'\(\d{4}\)', '', text)
df['Nombre de la Película'] = df['Nombre de la Película'].apply(limpiar_año)
```

La columna 'Nombre de la Película' contenía el año entre paréntesis. Para evitar redundancias con la columna 'Año', he realizado una limpieza utilizando la expresión regular (regex), que permite llevar a cabo operaciones de búsqueda y reemplazo para la limpieza de datos.

```python
#Ejercicio 3:Muestra tu creatividad.¿Qué crees relevante mostrar del Dataset "movies.dat" del ejercicio anterior?
#Realiza una o dos representaciones gráficas y justifica tu elección.


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/BIDDL3/PRE-PROCESSING-DATA/main/movies.dat"
df = pd.read_csv(url, sep='::', engine='python', names=['ID', 'Nombre de la Película', 'Género'], encoding='ISO-8859-1')

df['Decadas'] = df['Nombre de la Película'].str.extract(r'\((\d{4})\)').astype(float) // 10 * 10 

df = df.assign(Género=df['Género'].str.split('|')).explode('Género')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Decadas', y='Género', data=df, palette='Set3')
plt.xlabel('Década')
plt.ylabel('Género')
plt.title('Distribución de Géneros de Películas por Décadas')
plt.xticks(rotation=45)
plt.show()

```

En este gráfico, podemos observar cómo se representa la distribución de un género cinematográfico específico en una década en particular. La comedia, la aventura, la fantasía, el drama, el romance, el crimen y el thriller tuvieron su época de mayor distribución en las décadas de los años 80 y 90.

Por otro lado, notamos que los géneros épicos o de guerra, musicales, misterio, film-noir y western experimentaron una disminución en su distribución en el último siglo.

El horror, la ciencia ficción, la animación y el cine para niños surgieron entre las décadas de los años 70 y 90. Solo el género documental muestra valores atípicos en las épocas analizadas. Esto puede deberse a la falta de datos suficientes para realizar un análisis más específico o a que las películas de este género tienen valores significativamente alejados de la mediana. Esto podría indicar que el género tiende a producir películas que difieren notablemente en términos de distribución.

En resumen, basándonos en el análisis del gráfico de caja y bigotes, parece que estos géneros experimentaron diferentes niveles de distribución en distintas décadas.
```python
import pandas as pd
import matplotlib.pyplot as plt
import requests

url = "https://raw.githubusercontent.com/BIDDL3/PRE-PROCESSING-DATA/main/movies.dat"
df = pd.read_csv(url, sep='::', engine='python', names=['ID', 'Nombre de la Película', 'Género'], encoding='ISO-8859-1')

df['Año'] = df['Nombre de la Película'].str.extract(r'\((\d{4})\)').astype(float)
df_90s_2000s = df[(df['Año'] >= 1990) & (df['Año'] <= 2000)]

generos = df_90s_2000s['Género'].str.split('|').explode()
conteo_generos = generos.value_counts()

plt.figure(figsize=(12, 6))
conteo_generos.plot(kind='bar', color='chartreuse')
plt.title('Géneros más populares en los años 90 y 2000')
plt.xlabel('Género')
plt.ylabel('Cantidad de Películas')
plt.xticks(rotation=45)

plt.show()


```

A través de este gráfico, podemos observar que los géneros más populares en las décadas de los 90 y 2000 son el Drama y la Comedia. Esto podría indicar una preferencia general del público durante ese período. El gráfico también muestra géneros menos comunes que tuvieron una representación menor en las películas de esa época, como: El musical, el western y el film-noir. Esto podría ayudar a identificar nichos o géneros menos populares en comparación con los principales. A partir de este gráfico, podríamos profundizar en el análisis.
