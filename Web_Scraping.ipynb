#### Web Scraping con BeautifulSoup 

 http://quotes.toscrape.com:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
import scrapy
import warnings
from time import sleep
```

```python
# URL de la página
url = "http://quotes.toscrape.com"

# Realizar la solicitud a la página
response = requests.get(url)

# Verificar que la solicitud fue exitosa
if response.status_code == 200:
    # Crear una sopa (BeautifulSoup object) con el contenido de la respuesta
    soup = BeautifulSoup(response.text, 'html.parser')

    # Obtener citas y autores
    quotes = soup.find_all('span', class_='text')
    authors = soup.find_all('small', class_='author')

    for quote, author in zip(quotes, authors):
        print(f"{author.text}: {quote.text}")
else:
    print(f"Error al realizar la solicitud. Código de estado: {response.status_code}")
```

#### Web Scraping con Selenium

http://quotes.toscrape.com

```python


# URL de la página
url = "http://quotes.toscrape.com"

# Configurar Selenium para usar Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(url)

# Esperar hasta que la página se cargue completamente
time.sleep(3)

# Localizar las citas
quotes_elements = driver.find_elements(By.CLASS_NAME, 'text')
# Extraer el texto de las citas
quotes_text = [quote.text for quote in quotes_elements]

# Localizar los autores
authors_elements = driver.find_elements(By.CLASS_NAME, 'author')
# Extraer el texto de los autores
authors_text = [author.text for author in authors_elements]

# Cierra el navegador
driver.quit()

# Crear un DataFrame con las citas y los autores
quotes_df = pd.DataFrame({'Quote': quotes_text, 'Author': authors_text})

# Mostrar el DataFrame
print(quotes_df)

#Guardar el DataFrame en un archivo CSV
quotes_df.to_csv('quotes.csv', index=False)
```

https://es.wikipedia.org/wiki/Ada_Lovelace

```python
# Configurar Selenium para usar Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://es.wikipedia.org/wiki/Ada_Lovelace")

# Esperar a que se cargue la página
driver.implicitly_wait(10)

# Extraer los títulos de las secciones primera, segunda y tercera
section_titles = driver.find_elements(By.XPATH, "//h2/span[@class='mw-headline']")
first_section_title = section_titles[0].text
second_section_title = section_titles[1].text
third_section_title = section_titles[2].text

# Extraer los párrafos de las secciones primera, segunda y tercera
first_section_paragraphs = driver.find_elements(By.XPATH, "(//h2[1]/following-sibling::p)")
second_section_paragraphs = driver.find_elements(By.XPATH, "(//h2[2]/following-sibling::p)")
third_section_paragraphs = driver.find_elements(By.XPATH, "(//h2[3]/following-sibling::p)")
# Preparar los datos para el DataFrame
data = [
    {
        "Section Title": first_section_title,
        "Content": ' '.join([p.text for p in first_section_paragraphs])
    },
    {
        "Section Title": second_section_title,
        "Content": ' '.join([p.text for p in second_section_paragraphs])
    },
    {
        "Section Title": third_section_title,
        "Content": ' '.join([p.text for p in third_section_paragraphs])
    }
]

# Cierra el navegador
driver.quit()
# Conviertir la lista de diccionarios en un DataFrame
df_Ada = pd.DataFrame(data)

# Mostrar el DataFrame
print(df_Ada)

#Guardar el DataFrame en un archivo CSV
df_Ada.to_csv('Ada_Lovelace.csv', index=False)
```

### Ejercicio 3: Elige una página web que quieras y realiza web scraping mediante la librería Selenium primero y Scrapy después. 

la página web selecionada es: https://www.filmaffinity.com. Es un sitio web dedicado a la recomendación de películas basado en el concepto de "Almas Gemelas de Cine". Ofrece a los usuarios la posibilidad de calificar películas y series de televisión, así como de consultar reseñas, tráilers y rankings de popularidad. Además, el sitio proporciona información sobre estrenos en cines y lanzamientos en plataformas de streaming. Está diseñado para amantes del cine y no está vinculado a ninguna corporación o medio de comunicación.

Se hará web scraping de las Mejores series y películas de la historia del cine.

#### Selenium

```python


opts = Options()
#opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36")
opts.add_argument("--headless")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=opts    
)

driver.get("https://www.filmaffinity.com/es/topgen.php?genre=TV_SE&fromyear=&toyear=2019&country=&nodoc")

sleep(3)

contenedores_peliculas = driver.find_elements(By.XPATH, '//div[@class="mc-info-container"]')  
votaciones = driver.find_elements(By.XPATH, '//div[@class="avg-rating"]')

# Asegurarse de que cada película tiene una valoración correspondiente
if len(contenedores_peliculas) != len(votaciones):  
    print("Error: El número de películas no coincide con el número de valoraciones.")
else:
    peliculas = []

    for i in range(len(contenedores_peliculas)):
        titulo = contenedores_peliculas[i].find_element(By.XPATH, './/div[@class="mc-title"]').text
        pais = contenedores_peliculas[i].find_element(By.XPATH, './/img[@class="nflag"]').get_attribute('alt')
        director = contenedores_peliculas[i].find_element(By.XPATH, './/div[@class="mc-director"]').text
        valoracion = votaciones[i].text

        peliculas.append({
            "película": titulo,
            "país": pais,
            "director": director,
            "valoración": valoracion
        })

    driver.close()

    # Imprimir la información recolectada en formato de tabla
    for pelicula in peliculas:
        print(f'Título: {pelicula["película"]}, País: {pelicula["país"]}, Director: {pelicula["director"]}, Valoración: {pelicula["valoración"]}')
```

```python
# Crear un DataFrame a partir de la lista de diccionarios
df_peliculas_selenium = pd.DataFrame(peliculas)

# Guardar el DataFrame en un archivo CSV
df_peliculas_selenium .to_csv('df_peliculas_selenium', index=False)

# Imprimir el DataFrame
df_peliculas_selenium.head(30)
```

#### Scrapy

```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy import signals
import pandas as pd

class MiSpider(scrapy.Spider):
    name = 'mi_spider'  
    start_urls = ['https://www.filmaffinity.com/es/topgen.php?genres=&chv=0&orderby=avg&movietype=movie%7C&country=&fromyear=1972&toyear=1997&ratingcount=3&runtimemin=0&runtimemax=4']

    datos_raspados = []  # Lista para almacenar los datos raspados

    def parse(self, response):
        # Extraer todos los contenedores de información de películas
        contenedores_peliculas = response.xpath('//div[@class="mc-info-container"]')
        valoraciones = response.xpath('//div[@class="avg-rating"]')

        # Iterar a través de cada contenedor de película
        for i, contenedor in enumerate(contenedores_peliculas):
            pelicula = contenedor.xpath('.//div[@class="mc-title"]/a/text()').extract_first()
            pais = contenedor.xpath('.//div[@class="mc-title"]/img/@alt').extract_first()
            director = contenedor.xpath('.//div[@class="mc-director"]/div/span/a/text()').extract_first()
            valoracion = valoraciones[i].xpath('.//text()').extract_first()

            # Almacenar los datos extraídos
            datos_pelicula = {
                'película': pelicula,
                'país': pais,
                'director': director,
                'valoración': valoracion
            }
            self.datos_raspados.append(datos_pelicula)

    @classmethod #definir metodo perteneciente a "MiSpider"
    def from_crawler(cls, crawler, *args, **kwargs):
        spider = super(MiSpider, cls).from_crawler(crawler, *args, **kwargs)
        crawler.signals.connect(spider.spider_closed, signal=signals.spider_closed)
        return spider

    def spider_closed(self, spider):
      
        df_peliculas_scrapy = pd.DataFrame(self.datos_raspados)
        print(df_peliculas_scrapy)  # Imprimir el DataFrame
        # Guardar el DataFrame en un archivo CSV
        df_peliculas_scrapy.to_csv('peliculas_scrapy.csv', index=False, encoding='utf-8-sig')

# Configura y ejecuta el proceso de Scrapy
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/5.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(MiSpider)
process.start()
```
