```python
#2.3: Crea nuevas columnas (velocidad media del vuelo, si ha llegado tarde o no...)

import pandas as pd

df = pd.read_csv("C:/Users/acbon/Desktop/Data Science/DelayedFlights.csv")
df_reducido = df.sample(n=200000, random_state=random_seed)
random_seed = 42
df_reducido['VelocidadMedia'] = df_reducido['Distance'] / (df_reducido['AirTime']/60)
df_reducido['LlegadaTarde'] = df_reducido['ArrDelay'] > 15

print("Columna 'VelocidadMedia':")
print(df_reducido['VelocidadMedia'])

print("\nColumna 'LlegadaTarde':")
print(df_reducido['LlegadaTarde'])

```

```python
#2.4: Haz una tabla de las aerolíneas con mayores retrasos acumulados.

import pandas as pd

df = pd.read_csv("C:/Users/acbon/Desktop/Data Science/DelayedFlights.csv")
df_reducido = df.sample(n=200000, random_state=random_seed)
random_seed = 42

retrasos_acumulados = df.groupby('UniqueCarrier')['ArrDelay'].sum()

aerolineas_con_retraso = retrasos_acumulados.sort_values(ascending=False)

tabla_retrasos_acumulados = pd.DataFrame({'Aerolínea': aerolineas_con_retraso.index, 'Retraso Acumulado': aerolineas_con_retraso.values})

print(tabla_retrasos_acumulados)
```

```python
#2.5: ¿Cuáles son los vuelos más largos? ¿Y los más atrasados? Busca las rutas más largas y las que acumulen más retrasos.

import pandas as pd

df = pd.read_csv("C:/Users/acbon/Desktop/Data Science/DelayedFlights.csv")
df_reducido = df.sample(n=200000, random_state=random_seed)
random_seed = 42

rutas_mas_largas = df.groupby(['Origin', 'Dest'])['Distance'].sum()
rutas_mas_largas = rutas_mas_largas.sort_values(ascending=False)


retrasos_acumulados = df.groupby(['Origin', 'Dest'])['ArrDelay'].sum()
rutas_con_mas_retrasos = retrasos_acumulados.sort_values(ascending=False)

print("Rutas más largas:")
print(rutas_mas_largas.head(10))

print("\nRutas con más retrasos acumulados:")
print(rutas_con_mas_retrasos.head(10))
```
